```python
import cv2 as cv
import matplotlib.pyplot as plt
import numpy as np
```

```python
def readImage(url):
    img=cv.cvtColor(cv.imread(url),cv.COLOR_BGR2RGB)
    return img
    
```

```python
img=readImage("xian.jpg")
```

```python
plt.imshow(img)
```

```python
def shift(img):
    M=np.float32([[1,0,250],[0,1,500]])
    img=cv.warpAffine(img,M,(img.shape[0],img.shape[1]))
    return img
```

```python
plt.imshow(shift(img))
```

```python
def rotation(img):
    M=cv.getRotationMatrix2D((img.shape[0]//2,img.shape[1]//2),45,0.5)
    img=cv.warpAffine(img,M,(img.shape[0],img.shape[1]))
    return img
```

```python
plt.imshow(rotation(img))
```

```python
#interpolation=
# INTER_NEAREST
# INTER_LINEAR
# INTER_AREA
# INTER_CUBIC
# INTER_LANCZOS4
def resize(img):
    img=cv.resize(img,(255,255),interpolation=cv.INTER_CUBIC)
    return img
```

```python
plt.imshow(resize(img))
```
