```python
import numpy as np
import cv2 as cv
import matplotlib.pyplot as plt
def read(url):
    img= cv.cvtColor(cv.imread(url),cv.COLOR_BGR2RGB)
    plt.imshow(img)
    plt.axis("off")
    return img
img=read("face.jpg")
plt.imshow(img)
```

```python
kernelsizes=[(3,3),(9,9),(15,15)]
plt.figure(figsize=(15,15))
for i,kernel in enumerate(kernelsizes):
    plt.subplot(1,3,i+1)
    blur=cv.blur(img,kernel)
    plt.axis("off")
    plt.title("Blurred "+str(kernel))
    plt.imshow(blur)
    
plt.show()    
```

```python
kernelsizes=[(3,3),(9,9),(15,15)]
plt.figure(figsize=(15,15))
for i,kernel in enumerate(kernelsizes):
    plt.subplot(1,3,i+1)
    blur=cv.GaussianBlur(img,kernel,0)
    plt.axis("off")
    plt.title("GaussianBlur "+str(kernel))
    plt.imshow(blur)
    
plt.show()   
```

```python

plt.figure(figsize=(15,15))
for i,kernel in enumerate((3,9,15)):
    plt.subplot(1,3,i+1)
    blur=cv.medianBlur(img,kernel)
    plt.axis("off")
    plt.title("medianBlur"+str(kernel))
    plt.imshow(blur)
    
plt.show()   
```

```python

#作用是在保持边界清晰的情况下有效去除噪音
params=[(11,23,32),(12,11,24),(33,12,45)]
plt.figure(figsize=(15,15))
for i,(diameter,sigmaColor,sigmaSpace) in enumerate(params):
    plt.subplot(1,3,i+1)
    bilateral=cv.bilateralFilter(img,diameter,sigmaColor,sigmaSpace)
    plt.axis("off")
    plt.title("bilateral"+str(diameter )+" "+str(sigmaColor )+" "+str(sigmaSpace ))
    plt.imshow(bilateral)
    
plt.show()
```
