```python
import numpy as np
import cv2 as cv
import matplotlib.pyplot as plt
def read(url):
    img= cv.cvtColor(cv.imread(url),cv.COLOR_BGR2RGB)
    plt.imshow(img)
    plt.axis("off")
    return img
img=read("face.jpg")
plt.imshow(img)
```

```python
kernel=cv.getStructuringElement(cv.MORPH_ELLIPSE,(5,5))
erosion=cv.erode(img,kernel,iterations=4)
plt.imshow(erosion)
plt.show
```

```python
kernel=cv.getStructuringElement(cv.MORPH_ELLIPSE,(5,5))
dilate=cv.dilate(img,kernel,iterations=4)
plt.imshow(dilate)
plt.show
```

```python
kernel=cv.getStructuringElement(cv.MORPH_ELLIPSE,(5,5))
erosion=cv.morphologyEx(img,cv.MORPH_OPEN,kernel)
plt.imshow(erosion)
plt.show
```
