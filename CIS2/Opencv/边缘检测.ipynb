```python
import numpy as np
import cv2 as cv
import matplotlib.pyplot as plt
def read(url):
    img= cv.cvtColor(cv.imread(url),cv.COLOR_BGR2RGB)
    plt.imshow(img)
    plt.axis("off")
    return img
img=read("face.jpg")
plt.imshow(img)
```

```python
imggray=cv.cvtColor(img,cv.COLOR_RGB2GRAY)
plt.imshow(imggray,"gray")
```

```python
edges=cv.Canny(img,100,150)
plt.imshow(edges,"gray")
```

```python
def autoCanny(img,sigma=0.33):
    v=np.median(img)
    lower=int(max(0,(1.0-sigma)*v))
    upper=int(min(0,(1.0+sigma)*v))
    edged=cv.Canny(img,lower,upper)
    return edged

ed1=autoCanny(img,0.33)
plt.imshow(ed1)
```
