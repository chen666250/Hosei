```python
import numpy as np
import cv2 as cv
import matplotlib.pyplot as plt
```

```python
def read(url):
    img= cv.cvtColor(cv.imread(url),cv.COLOR_BGR2RGB)
    plt.imshow(img)
    plt.axis("off")
    return img
img=read("face.jpg")
```

```python
def pyrDown(img):
    imgc=img.copy()
    for i in range(5):
        imgc= cv.pyrDown(imgc)
        print(imgc.shape)
        plt.imshow(imgc)
        plt.show()

pyrDown(img)
```
