```python
import matplotlib.pyplot as plt
import numpy as np
import cv2 as cv
```

```python
def read(url):
    img= cv.cvtColor(cv.imread(url),cv.COLOR_BGR2RGB)
    plt.imshow(img)
    plt.axis("off")
    return img
img=read("face.jpg")
```

```python
green=(0,255,0)
```

```python
img=cv.line(img,(0,0),(400,400),green,5)
```

```python
plt.imshow(img)
```

```python
img=cv.rectangle(img,(0,0),(22,33),green,4)
```

```python
plt.imshow(img)
```

```python
img=cv.rectangle(img,(123,123),(23,333),green,-1)
plt.imshow(img)
```

```python
image1=np.zeros((300,300,3),dtype='uint8')
for i in range(10):
    radius=np.random.randint(5,200)
    color=np.random.randint(0,255,size=(3,)).tolist()
    pt=np.random.randint(0,300,size=(2,))
    cv.circle(image1,tuple(pt),radius,color,-1)
plt.imshow(image1)
```
