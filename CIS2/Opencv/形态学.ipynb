```python
import numpy as np
import cv2 as cv
import matplotlib.pyplot as plt
def read(url):
    img= cv.cvtColor(cv.imread(url),cv.COLOR_BGR2RGB)
    plt.imshow(img)
    plt.axis("off")
    return img
img=read("face.jpg")
plt.imshow(img)
```

```python
kernel=cv.getStructuringElement(cv.MORPH_ELLIPSE,(5,5))
erosion=cv.erode(img,kernel,iterations=4)
plt.imshow(erosion)
plt.show
```

```python
kernel=cv.getStructuringElement(cv.MORPH_ELLIPSE,(5,5))
dilate=cv.dilate(img,kernel,iterations=4)
plt.imshow(dilate)
plt.show
```

```python
kernel=cv.getStructuringElement(cv.MORPH_ELLIPSE,(5,5))
opening=cv.morphologyEx(img,cv.MORPH_OPEN,kernel)
plt.imshow(opening)
plt.show
```

```python
kernel=cv.getStructuringElement(cv.MORPH_ELLIPSE,(5,5))
closing=cv.morphologyEx(img,cv.MORPH_CLOSE,kernel)
plt.imshow(closing)
plt.show
```

```python
opening=cv.morphologyEx(img,cv.MORPH_OPEN,kernel)
closing=cv.morphologyEx(opening,cv.MORPH_CLOSE,kernel)
plt.imshow(closing)
```

#->-


```python
dilate=cv.dilate(img,kernel,iterations=4)
erosion=cv.erode(img,kernel,iterations=4)
plt.imshow(cv.subtract(dilate,erosion))
```

```python
gradient=cv.morphologyEx(opening,cv.MORPH_GRADIENT,kernel)
plt.imshow(gradient)
```

```python
what=cv.morphologyEx(img,cv.MORPH_TOPHAT,kernel)
plt.imshow(what)
```

```python
bhat=cv.morphologyEx(img,cv.MORPH_BLACKHAT,kernel)
plt.imshow(bhat)
```

```python
plt.imshow(cv.add(what,bhat))
```
