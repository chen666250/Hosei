```python
import numpy as np
import cv2 as cv
import matplotlib.pyplot as plt
def read(url):
    img= cv.cvtColor(cv.imread(url),cv.COLOR_BGR2RGB)
    plt.imshow(img)
    plt.axis("off")
    return img
img=read("face.jpg")
plt.imshow(img)
```

```python
imggray=cv.cvtColor(img,cv.COLOR_RGB2GRAY)
plt.imshow(imggray,"gray")
```

```python
ret1,thresh1=cv.threshold(imggray,127,255,cv.THRESH_BINARY)#127 阈值 255最大值
ret2,thresh2=cv.threshold(imggray,127,255,cv.THRESH_BINARY_INV)#127 阈值 255最大值
ret3,thresh3=cv.threshold(imggray,127,255,cv.THRESH_TRUNC)#127 阈值 255最大值
ret4,thresh4=cv.threshold(imggray,127,255,cv.THRESH_TOZERO)#127 阈值 255最大值
ret5,thresh5=cv.threshold(imggray,127,255,cv.THRESH_TOZERO_INV)#127 阈值 255最大值
```

```python
titles=['src','Binary','bI','T',"to","toi"]
images=[imggray,thresh1,thresh2,thresh3,thresh4,thresh5]
plt.figure(figsize=(15,5))
for i in range(6):
    plt.subplot(3,2,i+1)
    plt.imshow(images[i],"gray")
    plt.title(titles[i])
    plt.axis('off')
    plt.show()
    

plt.imshow(cv.bitwise_and(img,img,mask=thresh1))
```

```python
imggray=cv.medianBlur(imggray,5)
thresh7=cv.adaptiveThreshold(imggray,255,cv.ADAPTIVE_THRESH_MEAN_C,cv.THRESH_BINARY,11,3)
plt.imshow(thresh7,'gray')
```

```python

thresh8=cv.adaptiveThreshold(imggray,255,cv.ADAPTIVE_THRESH_GAUSSIAN_C,cv.THRESH_BINARY,11,3)
plt.imshow(thresh8,'gray')
```
