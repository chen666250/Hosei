```python
import cv2 as cv
import matplotlib.pyplot as plt
import numpy as np
```

```python
def read(url):
    img= cv.cvtColor(cv.imread("face.jpg"),cv.COLOR_BGR2RGB)
    plt.imshow(img)
    return img

img=read("face.jpg")
```

```python
detector=cv.CascadeClassifier("haarcascade_frontalface_alt.xml")
```

```python
rects=detector.detectMultiScale(img,scaleFactor=1.1,minNeighbors=2,minSize=(10,10),flags=cv.CASCADE_SCALE_IMAGE)

for (x,y,w,h) in rects:
    cv.rectangle(img,(x,y),(x+w,y+h),(0,255,0),2)
plt.imshow(img)
```
